```python
import pandas as pd
import numpy as np

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

sample = pd.read_csv('gender_submission.csv')
print(train.shape, test.shape)
```

```python
def engineer_features(df):
  #FamilySize
  df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
  df['Title'] = df['Name'].str.extract(r', ([A-Za-z]+)\.')
  df['Sex_encoded'] = df['Sex'].map({"male": 0, "female": 1})
  return df
```

```python
#Data cleaning
def data_clean(df):
 df.fillna({'Age':df['Age'].median()},inplace=True)
 df.dropna(subset='Embarked', inplace = True)
 df.drop(columns = 'Cabin', inplace=True)
```

```python
data_clean(train)
train.info()
```

```python
data_clean(test)
test.info()
```

```python
print(test.columns.tolist())
```

```python
# 1 missing title in test
train['Title'].fillna('Unknown', inplace=True)

#1 missing Fare in test
test['Fare'].fillna(test['Fare'].median(), inplace=True)
```

```python
# Features
features = ['Pclass','Age','Fare','Sex_encoded', 'FamilySize' ]

#define X_train, y_train
X_train = train[features]
y_train = train['Survived']

X_test_Kaggle = test[features]
```

```python
print(X_train.shape, y_train.shape, X_test_Kaggle.shape)
```

```python
# sklearn pipleine
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

#logistic Regression pipeline
lr_pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression())
])
dt_pipeline = Pipeline([
    ('model',DecisionTreeClassifier(max_depth=2, random_state=42))
])

```

```python
#train_test_split
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score,recall_score,f1_score,roc_auc_score
X_tr, X_val, y_tr, y_val = train_test_split(X_train,y_train, test_size=0.2, random_state=42)

#fit log REg
lr_pipeline.fit(X_tr, y_tr)

#predict
y_lr_pred = lr_pipeline.predict(X_val)

#accuracy
print(f"Accuracy of Logistic reg pipeline: {accuracy_score(y_val,y_lr_pred):.2f}")
print(f"precision of Logistic reg pipeline: {precision_score(y_val,y_lr_pred):.2f}")
print(f"recall of Logistic reg pipeline: {recall_score(y_val,y_lr_pred):.2f}")
print(f"f1 score of Logistic reg pipeline: {f1_score(y_val,y_lr_pred):.2f}")
print(f"AUC of Logistic reg pipeline: {roc_auc_score(y_val,y_lr_pred):.2f}")
```

```python
# fit Decision tree
dt_pipeline.fit(X_tr, y_tr)

#predict
y_dt_pred = dt_pipeline.predict(X_val)

#accuracy of decision tree pipeline
print(f"Accuracy of decision tree pipeline: {accuracy_score(y_val, y_dt_pred):.2f}")
print(f"precision of decision tree pipeline: {precision_score(y_val,y_dt_pred):.2f}")
print(f"recall of decision tree pipeline: {recall_score(y_val,y_dt_pred):.2f}")
print(f"f1 score of decision tree pipeline: {f1_score(y_val,y_dt_pred):.2f}")
print(f"AUC of decision tree pipeline: {roc_auc_score(y_val,y_dt_pred):.2f}")
```

```python
#Kaggle submission
final_predictions = dt_pipeline.predict(X_test_Kaggle)

submission = pd.DataFrame({
    'PassengerId': test['PassengerId'],
    'Survived': final_predictions
})
submission.to_csv('submission.csv', index=False)
print(submission.head())
```
