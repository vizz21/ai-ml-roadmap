1. Linear regression:

- A supervised learning algorithm used to predict output.
when the target variables is a continuous value, we call it regression.

- supervised learning algorithms take a number of inputs(features) and predict output y for the same.

- Linear regression uses the hypothesis which is a linear equation to predict output,
  h(x)= theta(0) + theta(1)X1 + theta(2)X2 ,
in case of predicting the house price (y), we have two inputs x1(house size),
x2(no. of rooms)

- Linear regression uses MSE to get the value of theta closer to output y.


#MSE: MSE = (1/n) × Σ(y_actual - y_predicted)²
- Average of the squared differences between predicted values and actual values. Measures how wrong your model is overall.
- Lower = better fit.


# Gradient descent:
- algorithm that minimises the MSE by repeatedly nudging m and b in the direction that reduces the error.
- Like walking downhill blindfolded, one small step at a time.


```python
import numpy as np
import pandas as pd
```

```python
# y =mx +b
# m,b -> direct formula

#gradient descent -> looks out for minimal MSE by changing m,b to reduce error

#direct formula may break down in large,complex datasets which require inverting a matrix, thus computationally harder
```

```python
#cost function (MSE)

# MSE = (1/n) × Σ(y_actual - y_predicted)²
def computed_mse(y_predicted,y_actual):
  #difference between each actual and predicted value
  diff = y_actual - y_predicted

  #square of differences
  square = np.square(diff)

  #mean of all squared differences
  result = np.mean(square)

  return result
```

```python
y_actual = np.array([3, 5, 7, 9])
y_predicted = np.array([2, 5, 8, 9])

#built-in function
computed_mse(y_predicted,y_actual)
```

```python
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_actual,y_predicted))
```

- learning rate is the way we choose to take baby steps or large steps, i.e. change values of theta accordingly,
-
- ((Alpha): The learning rate (or step size), which determines how large each step is.)
-----------------------------------------------------
1. if its too large -we oveeshoot our theta and get value (y_predicted) higher than y_actual,
2. if its small the algorithm is very slow and needs more iterations

## Gradient descent algorithm steps

- Prediction:

      1. y_predicted = m*X + b


- Compute the gradient:

      1. ∂MSE/∂m = (-2/n) × Σ x × (y_actual - y_predicted)
      2. ∂MSE/∂b = (-2/n) × Σ (y_actual - y_predicted)


> update rules below:

- Update Parameters (update rules): Change m and b to reduce the error:

      1. m = m - learning_rate × (∂MSE/∂m)
      2. b = b - learning_rate × (∂MSE/∂b)


```python
# gradient descent-> derive the update rule
def gradient_desc(X,y,learning_rate=0.01,iterations=10001):
  m=0
  b=0
  n=len(y)

  for _ in range(iterations):
  #1.Predict - using current m and b, calculate y_predicted for all X
    y_predict = m*X +b

    #2.Measure error - how wrong are the predictions vs actual y
  # calculate gradients
    dm = (-2/n) * (np.sum(X * (y-y_predict)))# gradient for m
    db = (-2/n) * (np.sum(y-y_predict))# gradient for b

    #3. Nudge - adjust m and b slightly to reduce that error

    ##update rules
    m = m - learning_rate * dm
    b = b - learning_rate * db

    mse = computed_mse(y, y_predict)
    if _ % 1000 == 0:
      print(f"Iteration {_}: MSE = {mse}")

  return m,b
```

```python
X=np.array([1,2,3,4,5])
y=np.array(([10,20,30,40,50]))

# scratch function

gradient_desc(X,y)
```

```python
#Sklearn linearRegression
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X.reshape(-1,1), y)

print(f"Sklearn slope: {model.coef_[0]}")
print(f"Sklearn intercept: {model.intercept_}")
```

```python
# MSE

def mse_scratch(y_actual1, y_predicted1):
  result = np.mean(np.square(y_actual1 - y_predicted1))

  return result
```

```python
# GRADIENT descent
def grad(X,y,alpha=0.01,iterations = 1000):
  m=0
  b=0
  n=len(y)

  for _ in range(iterations):
    #1. prediction - using current m,b calculate y_pred for all X
    y_pred = m * X + b





    #2. calculate gradients - for m,b
    dm = (-2/n) * (np.sum(X * (y - y_pred)))
    db = (-2/n) * (np.sum(y - y_pred))

    #3. update rule
    m = m - alpha * dm
    b = b - alpha * db

  return m,b
```

```python
X=np.array([1,2,3,4,5])
y=np.array(([10,20,30,40,50]))
grad(X,y)
#day 3
```
