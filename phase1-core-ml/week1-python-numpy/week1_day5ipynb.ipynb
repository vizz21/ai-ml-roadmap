  1. **cosine = (A.B) / ||A|| x ||B|**|

- Matrix multiply with @ -it is not the individual multiplication, rather the sum  of multiplication of respective rows X cols   
# - What does axis=0 vs axis=1 mean in NumPy? -axis-0 = rows,axis-1 = cols

```python
import numpy as np
```

```python
# day 4 recap
# cosine = (A.B) / ||A|| x ||B||
def cosine_similarity(v1, v2):
  magnitudeV1 = np.linalg.norm(v1)
  magnitudeV2 = np.linalg.norm(v2)

  cosine = (np.dot(v1,v2)) / (magnitudeV1 * magnitudeV2)
  return cosine

v1 = np.array([1,2,3])
v2 = np.array([4,5,6])
print(cosine_similarity(v1, v2))

#Matrix multiply with @
print(v1 @ v2)
```

## Normalisation
1. Scaling your data so every feature has the same range. Without it, a feature with values 0-10000 dominates a feature with values 0-1 in ML models

### The formula (z-score normalisation):
- ***normalised = (x - mean) / std***

     1.   height  weight  age
- person1  180    80      25
- person2  165    60      30
- person3  175    70      28

  2. Col-wise (most common in ML) → normalise each FEATURE separately. Height normalised against other heights, weight against weights.
  3. Row-wise → normalise each PERSON's values against each other. Less common.

```python
# Task 2: Write a normalisation function (row-wise, col-wise) using broadcasting only
        #Normalisation
def normalise(a, axis):
    # axis=0 → col-wise (normalise each column)
    # axis=1 → row-wise (normalise each row)

    # step 1: calculate mean along the axis
    mean = np.mean(a, axis,keepdims=True)

    # step 2: calculate std along the axis
    std = np.std(a, axis,keepdims=True)

    # step 3: return (a - mean) / std

    return (a-mean)/std

```

```python
#test array
w= np.arange(1,13,dtype=float).reshape(4,3)
print(w,"\n")

#calculate the column means
print(np.mean(w,axis=0,keepdims = True))
print(np.mean(w,axis=0),"\n")

#Subtract means from array:
result= (w - np.mean(w, axis=0,keepdims=True))
print(result,"\n")

#divide result by std
norm = result/np.std(w, axis=0, keepdims=True)
print(norm)
```

```python
#test normalisation function
normalise(w,axis=0)  #columns
```

```python
normalise(w,axis=1) #rows
```

```python
print(np.mean(normalise(w, axis=0), axis=0))  # should be ~0
print(np.mean(normalise(w, axis=1), axis=1))
```
