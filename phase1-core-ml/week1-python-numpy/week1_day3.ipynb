```python
def fibonacci():
  # initialise two variables
  first,next = 0,1

  # while true -->keeps the loping condition infinite in this case
  while True:
    #yield keyword for generator
    yield first

    #update
    first,next = next,next+first              # 0 1 1 2 3 5 8 13

a= fibonacci()
#print(next(a))
#print(__builtin__.next(a))
#print(__builtin__.next(a))
#print(__builtin__.next(a))
#print(__builtin__.next(a))
#print(__builtin__.next(a))

for _ in range(10):
  print(next(a))
```

```python
# FileManager class skeleton — just init, enter, exit
class FileManager():
  def __init__(self,filename,mode):
    self.filename = filename
    self.mode = mode

  #open("txt1","r")
  def __enter__(self):
    self.f = open(self.filename)

    return self.f

  def __exit__(self, exc_type, exc_val, exc_tb):
    # close self.f
    try:
      self.f.close()
    except Exception as e:
      print(f"Error: {e}")

```

# fancy indexing


> Fancy indexing = selecting specific rows or columns by passing a LIST of indices


1.   Get rows 0 and 2 only
2.   Get columns 1 and 3 only


1.   Get elements at positions (0,1), (2,3), (3,4) — specific individual elements






```python
bc= np.arange(1,21).reshape(4,5)
print(bc)
```

```python
# 1.   Get rows 0 and 2 only
bc[[0, 2]]
```

```python
#2.   Get columns 1 and 3 only
bc[:,[1,3]]
```

```python
# 3. Get elements at positions (0,1), (2,3), (3,4) — specific individual elements
print(bc[[0],[1]])
print(bc[2,3])         #indexing and slicing no fancy slicing
print(bc[[2],[3]])
print(bc[[3],[4]])

print(bc[[0,2,3],[1,3,4]],"fancy indexing")  #fancy indexing
```

# Task 3
- Broadcasting rules: understand why (3,1) + (1,4) = (3,4)

```python
#broadcast
d = np.arange(1,10).reshape(3,3)
print(d,d.shape)
```

```python
e = np.array([10,20,30])
print(e,e.shape)
```

```python
f = np.array([[10],[20],[30]])
print(f,f.shape)
```

```python
# (3,3) + (3,)
# d (3,3) + e(3,)   ->   e is treated like ([10],[20],[30])(1,3)
# then it is stretched to (3,3) so e becomes ([10,20,30],[10,20,30],[10,20,30])

print(d+e)
```

```python
# (3,3) + (3,1)
print(d+f)
```
