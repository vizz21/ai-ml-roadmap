```python
#Task 3:Implement cosine similarity between two vectors from scratch
#Cosine similarity measures how similar two vectors are. Used everywhere in ML — NLP, recommendation systems, search engines

#Formula:
'''
cos(θ) = (A · B) / (||A|| × ||B||)

'''
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

magnitudeA = np.square(v1)
print(magnitudeA)

magnitudeA_ = np.sqrt(np.sum(np.square(v1)))
print(magnitudeA_)


magnitudeB_ = np.sqrt(np.sum(np.square(v2)))
print(magnitudeB_)
```

```python
cosine = (np.dot(v1,v2))/(magnitudeA_ * magnitudeB_)
print(cosine)
```

```python
def cosine_similarity(v1, v2):
  magnitudeA_ = np.sqrt(np.sum(np.square(v1)))
  magnitudeB_ = np.sqrt(np.sum(np.square(v2)))

  cosine = (np.dot(v1,v2))/(magnitudeA_ * magnitudeB_)
  return cosine
```

```python
def cosine_similarity_(v1,v2):
  magnitudeA_ = np.linalg.norm(v1)
  magnitudeB_ = np.linalg.norm(v2)

  cosine = (np.dot(v1,v2))/(magnitudeA_ * magnitudeB_)
  return cosine
```
