```python
# Write a generator function that yields Fibonacci numbers
# 0 1 1 2 3 5 8 13 21

def fibonacci():
  prev,current = 0,1
  while True:
    yield prev
    prev,current = current,current + prev

fib = fibonacci()
for i in range(10):
  print(next(fib))
```

## Task 3 — OOP (30 min)
1. Build a DataLoader class that:

*  Takes a CSV filename in __init__
* Has a load() method that reads the file into a list of dicts
* Has a filter_rows() method that takes a column name and value, returns matching rows
* Has __len__ returning number of rows loaded
* Has __repr__ returning something like DataLoader(filename='data.csv', rows=150)





```python
import csv
class DataLoader():

  #Takes a CSV filename in init
  def __init__(self,filename) -> None:
    self.filename = filename
    self.data = []

  #load() method that reads the file into a list of dicts
  def load(self):
    with open(self.filename,"r") as dict1:
      reader = csv.DictReader(dict1)
      for row in reader:
          self.data.append(row)
  """
  Has a filter_rows() method that takes a column name and value,
  returns matching rows
  """
  def filter_rows(self,colName,value):
   return [i for i in self.data if i[colName] == value]

  def __len__(self):
      return (len(self.data))


  def __repr__(self):
    return f"Dataloader(filename = {self.filename},rows={len(self.data)})"

```

```python
loader = DataLoader("/content/sample_data/california_housing_train.csv")
loader.load()
print(loader)
print(len(loader))
```

```python
print(loader.filter_rows('longitude', 'latitude'))
print(loader.filter_rows('longitude', '-114.310000'))
```
