```python
# timer   → measures how long a function takes to run
import time



def my_decorator(func):
    def wrapper(*args, **kwargs):
        # record start time
        start = time.time()

        #run the original function
        result = func(*args, **kwargs)

          #record end time
        end = time.time()

          #print end - start
        print(end - start)
        return result
    return wrapper


@my_decorator
def sum(a,b):
  return a+b

print(sum(3,4))
```

### **except Exception as e** -> Python cleans up e when the block ends. Always save it to a separate variable if you need it later.

```python
# retry   → if a function fails/throws an error, try again up to 3 times
def retry_decorator1(func):

  def wrapper(*args, **kwargs):
    #logic
    last_error = None
    for i in range(3):
      try:
        return func(*args,**kwargs)
      except Exception as e:
        last_error = e
        print(f"Attempt {i+1} failed {e}")
    raise last_error
  return wrapper
```

```python
@retry_decorator1
def div(a,b):
  return a/b

div(0,1)
div(10,0)
```

```python
# exception handling — with open(), try/except/finally

""" practice
Opens a number from user input
Divides 100 by it
Handles ZeroDivisionError and ValueError separately
Uses finally to print "done"

"""


def exception_handling():


  try:
    num = int(input("Enter your number "))
    div = 100/num
    print(f"The result: {div}")
  except ZeroDivisionError as e:
    return (f"the typical zero error: {e}")
  except ValueError:
    print ("this is a value error :",ValueError)
  except Exception as e:
    return e
  finally:
    print("this is done")

exception_handling()
```

**Task 2: Context managers and exception handling — with open(), try/except/finally**

```python
# task 2: Context managers and exception handling — with open(), try/except/finally
"""

Task A — Write a function that reads a file safely:

Use with open()
Wrap it in try/except — handle FileNotFoundError specifically
Add a finally block that prints "attempt complete" whether it worked or not
Test it with a file that exists AND a filename that doesn't e
"""




def con(filename):
    #try except block
    try:
      #context manager 'with' helps  file handling
      #no need to use file.close() as we do if we use just file.open()
      with open(filename,"r") as f:
        print(f.read())
    #except block is the error block runs only if try block does not work
    except FileNotFoundError:
      print("Error : file not found")
    except Exception as e:
      print(f"Unexpected error: {e}")
    #finally block works everytime irresepective of try/except block outcome
    finally:
      print("attempt complete \n")

con("demo1.txt")
con("fake_file.txt")
con(123)
```

```python


#Task B — Write your own context manager using a class:

"""
__enter__ method — opens the file, returns it
__exit__ method — closes the file, handles any exception
Use it with with keyword

"""

class FileManager():

  def __init__(self,filename,mode):
    #store filename and mode
    self.filename = filename
    self.mode = mode

  def __enter__(self):
    #open the file, return it
    self.f = open(self.filename)

    return self.f

  def __exit__(self,exc_type, exc_val, exc_tb):
    #close the file, handle exception
    try:
      self.f.close()
    except exc_type as e:
      print(f"{Exception}")
    print(f"exc_type: {exc_type}")
    print(f"exc_val: {exc_val}")
    print(f"exc_tb: {exc_tb}")
    return True
with FileManager("demo1.txt", "r",) as f:
  print(f.read())

with FileManager("/content/demo1.txt", "r") as f:
    raise ValueError("test error")
```
