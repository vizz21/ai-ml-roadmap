```python
import numpy as np
```

```python
#1. mean
def my_mean(arr):
  #sum of all elements / number of elements
  mean = np.sum(arr) / len(arr)
  return mean
```

```python
arr = np.array([2, 4, 6, 8, 10])
print(my_mean(arr))        # my function
print(np.mean(arr))        # ground truth - must match
```

```python
# 2. variance
#VARIANCE shows how spread out our data is

#estimated variance = Avg. of the differences of squares between data and sample mean
#we take n-1 to make up for the difference of sample mean vs population mean

def my_variance(arr):
  variance = np.sum(np.square(arr -np.mean(arr))) / len(arr)
```

```python
def my_variance(arr,ddof=0):
    # step 1: calculate mean (use your my_mean function)
    mean = my_mean(arr)
    # step 2: calculate differences from mean
    diff = arr - mean
    # step 3: square the differences
    sq_diff = np.square(diff)

    # step 4: sum them up
    sum_variance = np.sum(sq_diff)

    # step 5:
    #divide by n (population variance)
    #divide by n-1 (sample variance)
    variance  = sum_variance/ (len(arr)-ddof)

    return variance
```

```python
arr = np.array([2, 4, 6, 8, 10])
print(my_variance(arr))    #my function
print(np.var(arr))         #numpy built in
```

```python
arr = np.array([2, 4, 6, 8, 10])
print(my_variance(arr,ddof=1))    #my function
print(np.var(arr,ddof=1))         #numpy built in
```

```python
#standard deviation :standard deviation is how much deviated the values are from mean
def my_std(arr,ddof=0):
  #square root of variance
  variance = my_variance(arr,ddof)

  std_ = np.sqrt(variance)
  return std_
```

```python
#3. covariance  -> how much two variables change together and whats the relation
# sum of (x - meanof X) * (y - mean of Y)/n-1
def my_covariance(x,y):
    #step 1: mean of x,y
    mean_X = my_mean(x)
    mean_Y = my_mean(y)

    #step 2: diff of variables with mean
    diff_X = x - mean_X
    diff_Y = y - mean_Y

    #Step 3: multiplye both differences
    multi = diff_X * diff_Y

    #step 4:
    cova = np.sum(multi) /  (len(x)-1)

    return cova
```

```python
x = np.array([2, 4, 6, 8, 10])   # e.g. hours studied
y = np.array([1, 3, 5, 7, 9])    # e.g. exam scores

print(my_covariance(x,y))
np.cov(x, y, ddof=1)[0][1]
```

```python
#4. Corelation : correlation(X,Y) = covariance(X,Y) / (std(X) * std(Y))

#covariance alone isnt enough to predict the next possible outcome
  # corelation determines whether the relation between two variables is strong,moderate or weak
  #thus a step closer to make our next prediction

  #we can use p-value depending on high/low confidence

def my_corelation(x,y):
  #covariance x,y
  cov = my_covariance(x,y)

  #step A if not using Standard devaition(Step B)
  #variance x,y
  var_x = my_variance(x,ddof=1)
  var_y = my_variance(y,ddof=1)
    #sqrt variance
  sqrt_var_x = np.sqrt(var_x)
  sqrt_var_y = np.sqrt(var_y)

  #step B if not using Standard devaition(Step A)
  #standard deviation of x,y
  std_x = my_std(x,ddof=1)
  std_y = my_std(y,ddof=1)



  # corelation
  corel =  cov / (std_x * std_y)
  #corel =  cov / (sqrt_var_x * sqrt_var_y)
  return corel



#standard deviation is just sq.rt of variance
```

```python
x = np.array([2, 4, 6, 8, 10])   # e.g. hours studied
y = np.array([1, 3, 5, 7, 9])    # e.g. exam scores
print(my_corelation(x,y))

print(np.corrcoef(x,y)[0][1])
```

```python
print(np.correlate(x,y))
```

```python
#5. Linear Regression
    # y = mx + b

# --> slope (m) = cov(X,Y) / var(X) -this is represented mathematically
                  #using the equation   --->>>>>
#--> intercept (b) = mean(Y) - m * mean(X)

def my_linearReg(x,y):
  #step 1: Calculate slope
   # m = covariance(x,y) / variance(x)
   m = my_covariance(x,y) / my_variance(x)

   #step 2: Calculate intercept
   b = np.mean(y) - (m * np.mean(x))


   return m,b
```

```python
my_linearReg(x,y)
```

```python
m, b = my_linearReg(x, y)

# predict y for a new x value
x_new = 12  # 12 hours studied
y_pred = m * x_new + b
print(f"Predicted exam score for {x_new} hours: {y_pred}")
```
