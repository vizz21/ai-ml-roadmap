```python
#Task 3.  Apply: apply(), map() — transform columns with custom functions
```

1. **map()** → works on a Series, maps each value to a new value
2. **apply()** → works on Series OR DataFrame, applies a function

        1. series.map() ->value substitution on a Series (what we just did with Sex)
        2. series.apply() ->  apply a function to each element of a Series




        1.   DataFrame.map() → apply a function elementwise to whole DataFrame
        2.   DataFrame.apply() → apply a function row-wise or column-wise




1. a. **Task — create an age group column:**
- Instead of raw ages, group them:

- 0-12 → "Child"
- 13-17 → "Teen"
- 18-60 → "Adult"
- 60+ → "Senior"

```python
#apply()
def get_age_group(age):
  if age <13:
    return "Child"

  elif age<18:
    return "Teen"

  elif age<61:
    return "Adult"

  else:
    return "Senior"
df['AgeGroup'] = df['Age'].apply(get_age_group)
print(df['AgeGroup'].value_counts())
print("\n",df['AgeGroup'])
```

```python
# Map is perfect for simple value substitutions.
#Like converting Sex column from "male"/"female" to 0/1 (which ML models need):

df['Sex_encoded'] = df['Sex'].map({'male':0,"female":1})
print(df['Sex_encoded'].value_counts())
```
