```python
#Task 1.a) cleaning

#1. drop rows only where Embarked is missing (only 2 rows, not worth filling)
df.dropna(subset='Embarked',inplace=True)
```

```python
print(df.info())
```

```python
from numpy import median
#2. fillna
# filling missing Age with the median age

# df['Age'].fillna(df['Age'].median(),inplace=True)  --> old syntax before pandas 3.0


df.fillna({'Age':df['Age'].median()},inplace=True)
# print(df['Age'].median())   #-->28.0
```

```python
#if duplicates drop them
df.drop_duplicates(inplace=True)
```

```python
#c) astype --> converting column data types.
print(df.dtypes)

# Survived  --> categorical variable -->0 or 1, dead or alive
df['Survived']= df['Survived'].astype(pd.CategoricalDtype())

print("Survived variable dtype changed: ",df['Survived'].dtype,"\n")

#Pclass category (1st, 2nd, 3rd class
df['Pclass'] = df['Pclass'].astype('category')
print("Pclass now converted into categorical variable")
print(df.dtypes)
```
