 ## Quick Week 1 recall
 1. Create a 4x5 NumPy array and slice the last two columns
 2. Boolean mask — values greater than 10
 3. Write my_mean() from scratch

```python
import numpy as np
```

```python
#  1. Create a 4x5 NumPy array and slice the last two columns
a = np.arange(1,21).reshape(4,5)
print(a)

#last 2 cols
print(a[:,-2:])
```

```python
#2. Boolean mask — values greater than 10
mask10 = a[a>10]
print(mask10)
```

```python
#3. Write my_mean() from scratch
def my_mean(arr):

  mean = np.sum(arr) / arr.size

  return mean
my_mean(a)
```

```python
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv')
print(df.shape)
```

	1. Pandas Series vs DataFrame: indexing with iloc, loc, boolean masking

```python
#dataframe - the whole table
print(type(df))

#series one column
print(type(df['Survived']))
```

```python
#indexing with iloc, loc

#iloc - indexing by position(number)
#first row
print(df.iloc[0])
print("\n-------------------\n")

## first 3 rows, first 3 cols
print(df.iloc[0:3,0:3])
```

```python
#loc - indexing by labels (column names)

# row-0 , col -Name
print(df.loc[0,'Name'])

print("\n-------------------\n")

# rows 0-2, specific columns
print(df.loc[0:2,['Name','Age']])

## loc → inclusive end → loc[0:2] gives rows 0,1,2
## iloc → exclusive end (like Python slicing) → iloc[0:3] gives rows 0,1,2
```

```python
# boolean masking

#1. Get all passengers who survived
#2. Get all female passengers over 30
#3. Get only the Name and Age columns for passengers in first class

#1. Get all passengers who survived
surv = df[df.Survived==1]
print(surv)
print("\n-------------------\n")



```

```python
#2. Get all female passengers over 30
female30 = df[(df.Sex=='female') & (df.Age>30)]
print(female30)
print("\n-------------------\n")
```

```python
#3. 3. Get only the Name and Age columns for passengers in first class
firstclass = df[df['Pclass']==1]
print(firstclass.loc[:,['Name','Age']])
```

```python
first = df.loc[df['Pclass']==1,['Name','Age']]
first
```
